```python
vids_2 = []
for (root, dirs, files) in os.walk('D:/yoga_train_valitation_1~10/'):
    for file in files:
        if file.endswith('.mp4'):
            vids_2.append({'root':root, 'dirs':dirs, 'files':file})
vids_df_2 = pd.DataFrame(vids_2)
vids_df_2.head()
```
