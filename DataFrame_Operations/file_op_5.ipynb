```python
import os
import pandas as pd
```

데이터프레임 행의 값은 다음을 포함합니다.

movement, val, difficulty, actor, take, camera, (color), video_dir

```python
root = 'D:/yoga_train_valitation_1~10/AI/Validation'
```

```python
def get_directory_table(root='D:/yoga_train_valitation_1~10/AI/Validation'):
    table = []
    movements = os.listdir(root)
    for movement in movements:
        movement_dir = os.path.join(root, movement)
        vals = os.listdir(movement_dir)
        for val in vals:
            val_dir = os.path.join(movement_dir, val)
            difficulties = os.listdir(val_dir)
            for difficulty in difficulties:
                difficulty_dir = os.path.join(val_dir, difficulty)
                actors = os.listdir(difficulty_dir)
                for actor in actors:
                    actor_dir = os.path.join(difficulty_dir, actor)
                    takes = os.listdir(actor_dir)
                    for take in takes:
                            take_dir = os.path.join(actor_dir, take)
                            cameras = os.listdir(take_dir)
                            for camera in cameras:
                                if camera.startswith('camera'):
                                    camera_dir = os.path.join(take_dir, camera)
                                    color_dir = os.path.join(camera_dir, 'color')
                                    if os.path.exists(color_dir):
                                        data = os.listdir(color_dir)
                                        for datum in data:
                                            if datum.endswith('.mp4'):
                                                video_dir = os.path.join(color_dir, datum)
                                                row = {'movement':movement, 'val':val, 'difficulty':difficulty, 'actor':actor, 'take':take, 'camera':camera, 'video_dir':video_dir}
                                                table.append(row)
    return table
```

```python
test = get_directory_table()
```

```python
def save_as_excel(excel_writer):
    vids_tb = get_directory_table()
    vids_df = pd.DataFrame(vids_tb)
    vids_df.to_excel(index=False, excel_writer=excel_writer)
```

```python
save_as_excel(excel_writer='../data/vids.xlsx')
```
