```python
my_root = r'D:\yoga_train_valitation_1~10'
my_root = os.path.normpath(my_root)
print(my_root)
```

```python
def get_walk_table(root=my_root): # 영상 데이터의 경로 정보를 얻어옵니다.
    table = []
    for (root, dirs, files) in os.walk(top=root):
        for file in files:
            if file.endswith('.mp4'):
                table.append({'root':root, 'dirs':dirs, 'file':file}) # 행 데이터
    return table
```

```python
table_1 = get_walk_table()
print(table_1[:10])
```

```python
testpath = table_1[0]['root']
testsplit = testpath.split(os.sep)
print(testsplit)
```

### DataFrame 생성 계획

한 행은 자세 명칭(pose), validation 폴더명(validation), 난이도(difficulty), 촬영 대상(subject), 촬영 정보(take), 카메라 번호(camera) 데이터로 구성됩니다.

```python
def get_path_table(walk_table):
    path_table = []
    for walk in walk_table:
        dir_split = walk['root'].split(os.sep)
        video_dir = os.path.join(walk['root'], walk['file'])
        row = {'pose':dir_split[4],
         'validation':dir_split[5],
         'difficulty':dir_split[6],
         'subject':dir_split[7],
         'take':dir_split[8],
         'camera':dir_split[9],
         'video':video_dir}
        path_table.append(row)
    return path_table
```

```python
my_data = get_path_table(table_1)
print(my_data[:5])
```

```python
my_df = pd.DataFrame(my_data)
my_df.head()
```

```python
def save_as_excel(root=r'D:\yoga_train_valitation_1~10', excel_writer=r'C:\Users\user\python\data\vids.xlsx'):
    walk_table = get_walk_table(root)
    path_table = get_path_table(walk_table)
    path_df = pd.DataFrame(path_table)
    path_df.to_excel(excel_writer=excel_writer, index=False)
```

```python
save_as_excel()
```
